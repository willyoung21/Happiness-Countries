```python
import pandas as pd

# Cargar el dataset combinado
df = pd.read_csv('../data/proccesed/final_happiness_data.csv')
```

Verificamos nulos y datos duplicados

```python
# Verificar valores nulos
print("Valores nulos en cada columna:\n", df.isnull().sum())

# Verificar duplicados
print("Filas duplicadas:", df.duplicated().sum())
```

visualizamos la distribución de variables numéricas para identificar patrones, sesgos o posibles anomalías.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogramas de las variables numéricas
df.hist(bins=20, figsize=(14, 10))
plt.show()

# Boxplots para detectar outliers
plt.figure(figsize=(14, 8))
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']))
plt.xticks(rotation=90)
plt.show()
```

```python
# Promedio de puntaje de felicidad por región
avg_happiness_region = df.groupby('Region')['Happiness_Score'].mean().sort_values()
print("Promedio de puntaje de felicidad por región:\n", avg_happiness_region)

# Visualización
plt.figure(figsize=(10, 6))
avg_happiness_region.plot(kind='bar', color='skyblue')
plt.ylabel("Promedio de Puntaje de Felicidad")
plt.title("Promedio de Puntaje de Felicidad por Región")
plt.show()
```
